Forma de uma Regresso Linear (equao de 1 grau)
# Y = b0 + b1 * X

- Declarando o modelo;
- n_jobs indica a quantidade de processos paralelos que vo calcular a regresso linear
- fit_intercept = False, indica que no h constante na equao criada; uso  raro!

```python
model = LinearRegression(n_jobs=None, fit_intercept=True)
```

- X: so as minhas variveis explicativas; tem que
- ser produzido como array bidimensional
- para acomodar mltiplas variveis por observao

```python
x = df2[['anos_exp','anos_estudo']].to_numpy()
```

- Y:  um array unidimensional, pois s h 1 varivel
- sendo prevista

```python
y = df2['salario'].to_numpy()
```

- faz a modelagem da regresso linear, calculando o
- intecepto e o coeficiente angular

```python
model.fit(x, y)
```

- score calcula o r-quadrado

```python
print(model.score(x, y))
```

- RLS: 0.9854136517127601
- RLM: 0.9868462044575527

- Complexidade de um algoritmo indica o tempo de processamento
- em funo da quantidade de dados
- RLS: O(n)
- RLM: O(n)

- R-quadrado varia de 0 at 1.
- 1  melhor possvel
- 0  pior possvel
