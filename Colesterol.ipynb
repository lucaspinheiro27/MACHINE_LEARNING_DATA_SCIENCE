```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import scipy.stats as stats

file = 'Colesterol.csv'
df = pd.read_csv(file)
df
```

```python
df.isna().sum()
```

```python
df1 = df
df_filtered = df1.query('Total.notnull() and HDL.notnull() and Idade.notnull()')
df_filtered
```

```python
df_filtered.isna().sum()
```

```python
x = df_filtered[['Total', 'HDL']]
y = df_filtered['Idade'].to_numpy()
model = LinearRegression()
model.fit(x, y)
score = model.score(x, y)
print(score)
```

```python
# Vou começar a reconstruir o HDL
# Vou utilizar como variável regressora o Total

df_filtered = df1[ (df1['Total'].notnull()) & (df1['HDL'].notnull()) & (df1['Idade'].notnull()) ]
x = df_filtered[['Total']].to_numpy()
y = df_filtered['HDL'].to_numpy()
model_hdl = LinearRegression()
model_hdl.fit(x, y)
score = model_hdl.score(x, y)
score
```

```python
def completar_hdl(linha):
    if pd.isna(linha['HDL']) and not pd.isna(linha['Total']) and not pd.isna(linha['Idade']):
        return round(model_hdl.predict([[linha['Total']]])[0])
    return linha['HDL']

# apply: executar a função chamada 1x para cada linha
df1['HDL_calculado'] = df1.apply(completar_hdl, axis=1)
df1
```

```python
df_filtered = df1[ (df1['Total'].notnull()) & (df1['Idade'].notnull()) & (df1['HDL_calculado'].notnull()) ]
x = df_filtered[['HDL_calculado']].to_numpy()
y = df_filtered['Total'].to_numpy()
model_total = LinearRegression()
model_total.fit(x, y)
score = model_total.score(x, y)
score
```

```python
def completar_total(linha):
    if pd.isna(linha['Total']) and not pd.isna(linha['Idade']) and not pd.isna(linha['HDL_calculado']):
        return round(model_total.predict([[linha['HDL_calculado']]])[0])
    return linha['Total']

# apply: executar a função chamada 1x para cada linha
df1['Total_calculado'] = df1.apply(completar_total, axis=1)
df1
```

```python
# Recalculo a minha regressão original utilizando as variáveis novas
x = df1[['Total_calculado', 'HDL_calculado']]
y = df1['Idade'].to_numpy()
model = LinearRegression()
model.fit(x, y)
score = model.score(x, y)
print(score)
```

* A regressão Linear sem o tratamento dos missings, obteve um r^2 de 0.15
* A diferença percentual no poder de explicação entre os dois coeficientes de determinação é aproximadamente 1.58%
* Já para regressão com os missings tratados, obteve uma significância menor com um r^2 de 0.14, mas sem viés e um poder de explicação maior
