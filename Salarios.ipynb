```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import scipy.stats as stats

df = pd.read_csv('Salarios.csv', decimal = ',')
df_renomeado = df.set_axis(['Anos_estudo', 'Anos_experiencia', 'Idade', 'Salario'], axis=1)
df = df_renomeado
df
```

```python
df.isna().sum()
```

```python
df1 = df
df_filtered = df1.query('Anos_estudo.notnull() and Anos_experiencia.notnull() and Idade.notnull() and Salario.notnull()')
df_filtered
```

```python
df_filtered.isna().sum()
```

```python
df_filtered.dtypes
```

```python
x = df_filtered[['Anos_estudo', 'Anos_experiencia', 'Idade']]
y = df_filtered['Salario'].to_numpy()
model = LinearRegression()
model.fit(x, y)
score = model.score(x, y)
print(score)
```

```python
# Vou começar a reconstruir Anos de estudo
# Vou utilizar como variáveis regressoras como Anos de experiência e Idade

df_filtered = df1[ (df1['Anos_estudo'].notnull()) & (df1['Anos_experiencia'].notnull()) & (df1['Idade'].notnull()) & (df1['Salario'].notnull()) ]
x = df_filtered[['Anos_experiencia', 'Idade']].to_numpy()
y = df_filtered['Anos_estudo'].to_numpy()
model_estudo = LinearRegression()
model_estudo.fit(x, y)
score = model_estudo.score(x, y)
score
```

```python
def completar_estudo(linha):
    if pd.isna(linha['Anos_estudo']) and not pd.isna(linha['Anos_experiencia']) and not pd.isna(linha['Idade']) and not pd.isna(linha['Salario']):
        campos = linha[['Anos_experiencia','Idade']]
        return round(model_estudo.predict([campos])[0])
    return linha['Anos_estudo']

# apply: executar a função chamada 1x para cada linha
df1['Anos_estudo_calculado'] = df1.apply(completar_estudo, axis=1)
df1
```

```python
# Vou começar a reconstruir Anos de experiência
# Vou utilizar como variáveis regressoras como Anos de estudo calculado e Idade

df_filtered = df1[ (df1['Anos_experiencia'].notnull()) & (df1['Idade'].notnull()) & (df1['Salario'].notnull()) & (df1['Anos_estudo_calculado'].notnull()) ]
x = df_filtered[['Idade', 'Anos_estudo_calculado']].to_numpy()
y = df_filtered['Anos_experiencia'].to_numpy()
model_experiencia = LinearRegression()
model_experiencia.fit(x, y)
score = model_experiencia.score(x, y)
score
```

```python
def completar_experiencia(linha):
    if pd.isna(linha['Anos_experiencia']) and not pd.isna(linha['Idade']) and not pd.isna(linha['Salario']) and not pd.isna(linha['Anos_estudo_calculado']):
        campos = linha[['Idade', 'Anos_estudo_calculado']]
        return round(model_experiencia.predict([campos])[0])
    return linha['Anos_experiencia']

# apply: executar a função chamada 1x para cada linha
df1['Anos_experiencia_calculado'] = df1.apply(completar_experiencia, axis=1)
df1
```

```python
df_filtered = df1[ (df1['Idade'].notnull()) & (df1['Salario'].notnull()) & (df1['Anos_estudo_calculado'].notnull()) & (df1['Anos_experiencia_calculado'].notnull()) ]
x = df_filtered[['Anos_estudo_calculado', 'Anos_experiencia_calculado']].to_numpy()
y = df_filtered['Idade'].to_numpy()
model_idade = LinearRegression()
model_idade.fit(x, y)
score = model_idade.score(x, y)
score
```

```python
def completar_idade(linha):
    if pd.isna(linha['Idade']) and not pd.isna(linha['Salario']) and not pd.isna(linha['Anos_estudo_calculado']) and not pd.isna(linha['Anos_experiencia_calculado']):
        campos = linha[['Anos_estudo_calculado', 'Anos_experiencia_calculado']]
        return round(model_idade.predict([campos])[0])
    return linha['Idade']

# apply: executar a função chamada 1x para cada linha
df1['Idade_calculado'] = df1.apply(completar_idade, axis=1)
df1
```

```python
# Recalculo a minha regressão original utilizando as variáveis novas
x = df1[['Anos_estudo_calculado', 'Anos_experiencia_calculado', 'Idade_calculado']]
y = df1['Salario'].to_numpy()
model = LinearRegression()
model.fit(x, y)
score = model.score(x, y)
print(score)
```

* A regressão Linear sem o tratamento dos missings, obteve um r^2 de 0.90
* Já para regressão com os missings tratados, obteve uma significância menor com um r^2 de 0.64, mas sem viés e um poder de explicação maior
* A diferença percentual no poder de explicação entre os dois coeficientes de determinação é aproximadamente 28.88%
* Sugerindo que o modelo ajustado com os dados faltantes tratados é mais robusto e generalizável, resultando em um modelo mais representativo
