```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import statsmodels.api as sm
```

```python
#df = pd.read_csv('runner.csv', sep = "\t")
#df = pd.read_csv('imc.csv', sep = "\t")
df = pd.read_csv('visits.csv', sep = "\t")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df['dow'] = df['date'].dt.dayofweek
```

```python
df['mon'] = df.apply(lambda row: 1 if row['dow']==0 else 0, axis=1) # monday
df['tue'] = df.apply(lambda row: 1 if row['dow']==1 else 0, axis=1) # tuesday
df['wed'] = df.apply(lambda row: 1 if row['dow']==2 else 0, axis=1) # wednesday
df['thu'] = df.apply(lambda row: 1 if row['dow']==3 else 0, axis=1) # thursday
df['fri'] = df.apply(lambda row: 1 if row['dow']==4 else 0, axis=1) # friday
df['sat'] = df.apply(lambda row: 1 if row['dow']==5 else 0, axis=1) # saturday
df['sun'] = df.apply(lambda row: 1 if row['dow']==6 else 0, axis=1) # sunday
```

```python
model = LinearRegression(n_jobs=None, fit_intercept=True, copy_X=True)
```

```python
x = df[['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']].to_numpy()
y = df['visits']
```

```python
model.fit(x, y)
model.score(x, y)
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
```
