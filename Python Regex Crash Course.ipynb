```python
import re
p = re.compile('([a-z]+\s?)', re.IGNORECASE)
p.findall('fernando nemec da silva')
```

```python
import re
p = re.compile('([a-z]+\s?)', re.IGNORECASE)
p.match('fernando nemec da silva')
```

```python
import re

s = """
fernando nemec
da silva
"""

p = re.compile('([a-z]+\s?)', re.IGNORECASE | re.MULTILINE)
p.findall(s)
```

```python
p = re.compile('(neme.+?silva)', re.IGNORECASE | re.MULTILINE)
p.findall(s)
```

```python
import re
s = '<button type="submit" class="btn">Send</button>'
matches = re.finditer('".+?"', s)
for match in matches:
    print(match.group())
```

```python
import re
s = '<button type="submit" class="btn">Send</button>'
matches = re.finditer('([^\s]*?)="(.*?)"', s)
for match in matches:
    print(match.group())
```

```python
import re
s = '<button type="submit" class="btn">Send</button>'
matches = re.finditer(' (([^\s]+?)="(.+?)")', s)
for match in matches:
    print(match.group())
    print(match.group(1))
    print(match.group(2))
    print(match.group(3))
```

```python
statement = 'Please contact us at: support@datacamp.com'
match = re.search(r'([\w\.-]+)@([\w\.-]+)', statement)
if statement:
  print("Email address:", match.group()) # The whole matched text
  print("Username:", match.group(1)) # The username (group 1)
  print("Host:", match.group(2)) # The host (group 2)
```
